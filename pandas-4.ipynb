### 4. feladat
- Telepítsül a plotly modult ([https://plotly.com/python/getting-started/#installation](https://plotly.com/python/getting-started/#installation)).
- Ábrázoljuk a vulkánok elhelyezkedését térképen a plotly segítségével.

```python
import plotly.express as px
```

```python
fig = px.scatter_geo(df, lat='Latitude',lon='Longitude', projection="natural earth")
fig.update_layout(title = 'Volcanoes', title_x=0.5)
fig.show()
```

```python
# Próbáljuk ki a 'color' és a 'hover_data' paramétereket:

fig = px.scatter_geo(df, lat='Latitude',lon='Longitude', color = 'Country', hover_data = ['Volcano Name', 'Elevation (m)'], projection="natural earth")
fig.update_layout(title = 'Volcanoes', title_x=0.5)
fig.show()
```
