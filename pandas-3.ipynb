### [Csoportosítás](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html)(Groupby)

Pandasban a csoportosítás folyamata az alábbi lépésekből áll:
- Az adatok felosztása (split) csoportokra, valamilyen feltétel alapján.
- Valamely függvény alkalmazása (apply) az egyes csoportokra külön-külön.
- Az eredmények kombinálása (combine) egy adatszerkezetbe.

Ezek közül a felosztás a legegyszerűbb. A felosztás kritériuma általában egy oszlopban (vagy oszlop kombinációban) található érték. A függvényalkalmazás lehet aggregáló jellegű (pl. csoportonkénti elemszám, összeg, átlag, minimum, maximum, első rekord, utolsó rekord) vagy egyéb (pl. csoportonkénti standardizálás, adathiány kitöltés vagy szűrés). A kombinálási lépés az aggregáló típusú függvények esetén automatikusan lefut, egyéb esetekben a programozónak kell kezdeményeznie.

```python
# Használjuk az előző DataFrame-et!
df
```

```python
# Csoportosítás tantárgyak szerint.
gb = df.groupby('tantárgy')
```

```python
# Az eredmény típusa.
type(gb)
```

```python
# Hány rekord tartozik az egyes tantárgyakhoz?
gb.size()
```

```python
# Ugyanez az átmeneti gb objektum használata nélkül.
df.groupby('tantárgy').size()
```

```python
# Tantárgyankénti átlag.
df.groupby('tantárgy')['osztályzat'].mean() # Csoportosítunk tantárgyak szerint, majd ezen az osztályzatokat vesszük. Ezen használjuk a mean-t
```

```python
# Tantárgyankénti átlag minden tanulóhoz.
df.groupby(['tanuló', 'tantárgy'])['osztályzat'].mean() # Vesszük a tanulókat, azokon belül a tantárgyakat, majd ezen az osztályzatokat.
```

```python
type(df.groupby(['tanuló', 'tantárgy'])['osztályzat'].mean())
```

```python
# Az index átalakítása két hagyományos oszloppá.
df.groupby(['tanuló', 'tantárgy'])['osztályzat'].mean().reset_index()
```

```python
# 1. Az alapvető információk (len, shape, .columns, .info(), .describe()):

len(pl)

pl.shape

pl.columns

pl.info()

pl.describe()
```

```python
# 2. Adjuk meg, hogy összesen hány hazai és hány vendég gól született. (.sum())

print(pl['hazai gól'].sum())

print(pl['vendég gól'].sum())

# print nélkül csak a legutolsót írja ki, ami nem elég
```

```python
# 3. Adjunk hozzá egy új oszlopot, amely a gólokat (hazai+vendég) tartalmazza.

pl['gólok'] = pl['hazai gól'] + pl['vendég gól']
pl
```

```python
# 4. Határozzuk meg, hogy meccsenként átlagosan hány gól született. (.mean())

pl['gólok'].mean()
```

```python
# 5. A mérkőzések hány százalékán esett gól?

len(pl[ pl['gólok'] > 0]) / len(pl) * 100 # Százalékszámítás: (volt gól / nem volt gól) * 100

# másképp:
(pl['gólok'] > 0).sum() / len(pl) *100
```

```python
# 6. Melyik mérkőzésen esett a legtöbb gól?

pl[ pl['gólok'] == pl['gólok'].max() ] # Ahol a gólok száma egyenlő a legtöbb góllal

# másképp:
pl.iloc[pl['gólok'].idxmax()]
```

```python
# 7. Tekintsük a Liverpool FC mérkőzéseit.
# A. Átlagosan hány gólt szereztek hazai pályán és hányat kaptak?
szerzett = pl[ pl['hazai csapat'] == 'Liverpool FC']['hazai gól'].mean()
kapott = pl[ pl['hazai csapat'] == 'Liverpool FC']['vendég gól'].mean()
print(f'Átlagosan szerzett: {szerzett}, átlagosan kapott: {kapott} ')

# B. Hány hazai meccset nyertek meg?
# megszámoljuk, hogy a Liverpool FC meccseinél hány olyan van, 
# ahol a hazai gólok száma nagyobb vendég gólok számánál.
(pl[ pl['hazai csapat'] == 'Liverpool FC']['hazai gól'] > pl[ pl['hazai csapat'] == 'Liverpool FC']['vendég gól']).sum()


# másképp: 
# a hazai győzelmek közül megszámoljuk, hogy hány olyan volt,
# amikor a Liverpool FC volt a hazai csapat
feltetel1 = (pl['hazai gól'] > pl['vendég gól'])
feltetel2 = (pl['hazai csapat'] == 'Liverpool FC')
len( pl[ feltetel1 & feltetel2] )

# C. Idegenben mennyi volt a legtöbb góljuk?
pl[ pl['vendégcsapat'] == 'Liverpool FC']['vendég gól'].max()
```

```python
# 8. Melyik fordulóban hány gól esett?
pl.groupby('forduló')['gólok'].sum()
```

```python
# 9. Melyik fordulóban esett a legtöbb gól?

pl.groupby('forduló')['gólok'].sum().idxmax()
```

```python
# 10. Melyik fordulóban mennyi volt az átlagos mérkőzésenkénti gólszám?

pl.groupby('forduló')['gólok'].mean()
```

```python
# 11. Írjuk ki a tíz leggólgazdagabb fordulót, gólszámokkal együtt!

pl.groupby('forduló')['gólok'].sum().sort_values(ascending=False)[:10] # ascending: növekvő
```

```python
# 12. Határozzuk meg, hogy a bajnokság végén melyik csapatnak hány pontja volt (győzelem 3 pont, döntetlen 1 pont, vereség 0 pont).

# Minden egyes meccsre kiszámítjuk, hogy az adott meccsen hány hazai és 
# hány vendégpont született. Majd minden egyes csapatra kiszámítjuk, hogy
# hány pontot szerzett hazai csapatként és mennyit vendégként. Végül ezek
# összege adja a végső pontszámot.
#
# A '.sort_values(ascending=False)' paranccsal a pontszámok szerint csökkenő 
# sorrendbe rendezünk. Enélkül ábécé szerinti sorrendet kapunk.
#


pl['hazai pont'] = ( (pl['hazai gól'] > pl['vendég gól'] ) *3 + 
                     (pl['hazai gól'] == pl['vendég gól'] ) *1 )

pl['idegenbeli pont'] = ( (pl['hazai gól'] < pl['vendég gól'] ) *3 + 
                          (pl['hazai gól'] == pl['vendég gól'] ) *1 )

pontszám = (pl.groupby('hazai csapat')['hazai pont'].sum() + 
            pl.groupby('vendégcsapat')['idegenbeli pont'].sum() 
           ).sort_values(ascending=False)

pontszám
```
