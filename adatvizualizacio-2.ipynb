```python
anscombe = sns.load_dataset('anscombe')
```

Példa:

Rajzoljuk ki a már ismert Anscombe adathalmazokat a regressziós egyenessel együtt az sns.lmplot használatával.

```python
sns.lmplot(
    data=anscombe, x="x", y="y", col="dataset", hue="dataset",
    col_wrap=4, palette="muted", ci=None,
    height=4, scatter_kws={"s": 100, "alpha": 1}
)
plt.suptitle('Anscombe', fontsize=16, fontweight='bold')
plt.tight_layout(h_pad=2)
plt.show()
```

10. Feladat:

Ábrázoljuk hőtérképen a numerikus változók közötti korrelációt: sns.heatmap.

```python
pcorr = penguins2.corr(numeric_only=True) # a .corr() függvény kiszámolja a numerikus változók közötti korrelációs együtthatókat, és egy mátrixot ad vissza, ahol a sorok és oszlopok a változókat jelölik, és az értékek a korrelációs együtthatók
pcorr
```

```python
sns.heatmap(pcorr)
plt.show()
```

```python
sns.heatmap(pcorr, annot=True, vmin=-1, vmax=1, cmap='Blues') # annot: a négyzetekbe írja ki a számokat, vmin és vmax: a színezés skálájának minimuma és maximuma, cmap: színpaletta
plt.show()
```

```python
# a korrelációs együttható abszolútértéke:
sns.heatmap(abs(pcorr), annot=True, vmin=0, vmax=1, cmap='Blues') # annot: a négyzetekbe írja ki a számokat, vmin és vmax: a színezés skálájának minimuma és maximuma, cmap: színpaletta
plt.show()
```
