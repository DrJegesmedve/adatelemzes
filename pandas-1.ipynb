
## t-próba és változatai
### 1. Egymintás t-próba

Feltételek:
- a minta normális eloszlásból származik
- az elméleti szórás nem ismert
- a nullhipotézis az ismeretlen várható értékre vonatkozik
- dokumentáció: [ttest_1samp](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_1samp.html)

```python
# Létrehozzuk a mintát, egy Numpy tömböt.
# Ez egy mintavételnek felel meg. 10 db adat, nem az összes!
data = np.array([0.8, 2.8, 0.1, -0.9, -0.6, -1.3, 0.5, -1.6, -1.1, -1.8]) # Numpy tömb
```

```python
print(len(data)) # Numpy tömb hossza, elemszáma
print(data.mean()) # Átlag
print(data.std()) # Szórás: Átlagtól való átlagos eltérés
```

```python
# Kétoldali t-próba
# Azért kétoldali, mert csak azt vizsgálja, hogy az átlag eltér-e a várható értéktől? Az mindegy, hogy lefelé vagy felfelé.
stats.ttest_1samp(data, -0.6) # Adatok, várható érték
# Ha a pvalue kevesebb, mint az 5%, akkor elutasítjuk
# EREDMÉNY:
# statistic (t-érték): Minél nagyobb, annál valószínűbb, hogy az állításunk hamis (abszolút érték)
# pvalue (p-érték): Ha az állításunk helyes, mekkora az esélye, hogy egy ilyen átlagú mintát fogtunk ki? (5% alatt elutasítjuk a nullhipotézist, 95%-os szignifikanciaszintnél)
# df (Szabadsági fok): Elemek száma - 1; 
```

```python
# Egyoldali t-próba
# Azt vizsgálja, hogy a várható érték kisebb vagy nagyobb, mint az átlag. Tehát a 0,6 esetén 96% bizonyossággal mondja.
# Ha a pvalue kevesebb, mint az 5%, akkor elutasítjuk
stats.ttest_1samp(data, 0.6, alternative='less') # Adatok, várható érték, kisebb/nagyobb
# alternative: less/greater/two-sided
# Mondatba foglalva: Az én gyanúm, hogy a 0,6 kisebb (less), mint a nullhipotézis. A pvalue értéke ezt mondja meg, hogy igaz-e.
```


### 2. Páros t-próba

- Párosított adatokra vonatkozó t-próba
- dokumentáció: [ttest_rel](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mstats.ttest_rel.html)

```python
# Ugyanazat megmérjük 2 adatsoron
data1 = np.array([0.8, 2.8, 0.1, -0.9, -0.6, -1.4, 0.5, -1.6, -1.1, -1.8])
data2 = np.array([1.1, -0.4, -0.9, -0.7, -0.8, -0.2, 0.5, 1.2, -1.2, -0.2])
```

```python
print( len(data1), data1.mean(), data1.std() )
print( len(data2), data2.mean(), data2.std() )
```

```python
stats.ttest_rel(data1, data2) # A két adat
# Nem tudjuk elutasítani, mert 75% -> pvalue
# A két adat sorrendje mindegy, csak a statistic előjele fordul meg.
```

```python
stats.ttest_rel(data1,data2, alternative='greater') # Adat1, adat2,
# FONTOS A SORREND EGYOLDALI T-PRÓBÁNÁL
# Ha a data1 szerintem kisebb (less): data1, data2 # term: data 1 is less than data2
# Ha a data1 szerintem kisebb (greater): data2, data1
```


### 3. Kétmintás t-próba

- Két független minta összehasonlítására vonatkozó t-próba
- dokumentáció: [ttest_ind](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html)

```python
stats.ttest_ind(data1, data2)
```

```python
stats.ttest_ind(data1, data2, equal_var=False)# equal_var: Variancia (szórásnégyzet)
# Student féle t-próba: True, a két csoportban hasonló nagyságú az ingadozás (variancia)
# Welch féle t-próba: False, ha az egyik csoport nagy szórású, a másik pedig kis szórású. Ez mindig pontosabb.
```

## Chí-négyzet próbák

- leírás: https://en.wikipedia.org/wiki/Chi-squared_test

### 7. Chí-négyzet próba illeszkedés vizsgálatra

- dokumentáció: [chisquare](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chisquare.html)

```python
megfigyelt = np.array([30, 35, 20, 10, 5])
elvart = np.array([25, 40, 25, 5, 5])
```

```python
stats.chisquare(megfigyelt, elvart)
# Kis p-value esetén (5% alatt) elutasítjuk
```

### 8. Chí-négyzet próba függtelenség vizsgálatra

- dokumentáció: [chi2_contingency](https://docs.scipy.org/doc/scipy/tutorial/stats/hypothesis_chi2_contingency.html)

```python
table = np.array([ [ 20, 30], [40, 15]])
```

```python
stats.chi2_contingency(table)
# Összefügg-e két dolog?
# statisctic: Mekkora a távolság a mérés és a független táblázat között?
# p-value: 5% alatt összefüggés van, azaz nem függetlenek. 
```
